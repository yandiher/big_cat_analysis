# import library and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = [
    'lion',
    'jaguar',
    'leopard',
    'tiger',
    'cheetah',
    'puma']
```

```python
body_height = [
    120,
    72,
    65,
    100,
    81,
    75]
```

```python
body_weight = [
    200,
    120,
    51,
    230,
    47,
    77]
```

```python
tail_long = [
    75,
    60,
    84,
    98,
    70,
    79]
```

```python
speed = [
    74,
    50,
    58,
    57,
    105,
    72]
```

```python
bite_pressure = [
    650,
    1500,
    305,
    1050,
    475,
    724]
```

```python
hunting_rate = [
    0.25,
    0.05,
    0.38,
    0.10,
    0.58,
    0.82]
```

```python
ability = [
    'roar',
    'swim',
    'climb',
    'mediocre',
    'run',
    'jump']
```

# make dataframe

```python
dataset = pd.DataFrame({
    'name': names,
    'body_height': body_height,
    'body_weight': body_weight,
    'tail_long': tail_long,
    'speed': speed,
    'bite_pressure': bite_pressure,
    'hunting_rate': hunting_rate,
    'ability': ability})
```

```python
dataset
```

```python
dataset['body_height'] = (dataset.body_height/dataset.body_height.max()).round(2)*100
dataset['body_weight'] = (dataset.body_weight/dataset.body_weight.max()).round(2)*100
dataset['tail_long'] = (dataset.tail_long/dataset.tail_long.max()).round(2)*100
dataset['speed'] = (dataset.speed/dataset.speed.max()).round(2)*100
dataset['bite_pressure'] = (dataset.bite_pressure/dataset.bite_pressure.max()).round(2)*100
dataset['hunting_rate'] = dataset.hunting_rate*100
dataset
```

```python
100+87+77+70+43+25
```

```python
dataset['aggregate'] = ((dataset.body_height+dataset.body_weight+dataset.tail_long+dataset.speed+dataset.bite_pressure+dataset.hunting_rate)/6).round(2)
```

```python
dataset
```

```python
for i in dataset.columns[1:7]:
    order_list = dataset.sort_values(by=i,ascending=False).name
    plt.figure(figsize=(15,5), dpi=300)
    ax = sns.barplot(data=dataset, x=i, y=dataset['name'], order=order_list)
    ax.bar_label(ax.containers[0])
    plt.title(i)
    plt.show()
```

```python
order_list = dataset.sort_values(by='aggregate',ascending=False).name
plt.figure(figsize=(15,5), dpi=300)
ax = sns.barplot(data=dataset, x='aggregate', y=dataset['name'], order=order_list)
ax.bar_label(ax.containers[0])
plt.title('aggregate')
plt.show()
```
